```python
import pandas as pd
from pathlib import Path

RAW_PATH = Path("../data/raw/microdados_enem_2023/DADOS/MICRODADOS_ENEM_2023.csv")
OUTPUT_PATH = Path("../data/interim/microdados_enem_2023_reduzido.csv")

colunas = [
    "NU_INSCRICAO",
    "NU_ANO",
    "TP_FAIXA_ETARIA",
    "TP_SEXO",
    "TP_COR_RACA",
    "SG_UF_PROVA",
    "TP_ESCOLA",
    "TP_PRESENCA_CN",
    "TP_PRESENCA_CH",
    "TP_PRESENCA_LC",
    "TP_PRESENCA_MT",
    "NU_NOTA_CN",
    "NU_NOTA_CH",
    "NU_NOTA_LC",
    "NU_NOTA_MT",
    "NU_NOTA_REDACAO",
    "Q006",
    "Q025"
]

df = pd.read_csv(
    RAW_PATH,
    sep=";",
    encoding="latin-1",
    usecols=colunas
)

df = df[colunas]

df.to_csv(
    OUTPUT_PATH,
    index=False,
    sep=";",
    encoding="utf-8"
)
```
